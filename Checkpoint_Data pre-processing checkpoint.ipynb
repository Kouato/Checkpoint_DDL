4) Inspectez l'ensemble de données pour détecter d'éventuelles valeurs manquantes.

```python
df.isnull().sum()
```

D'après notre DataFrame ici nous avons deux colonnes qui contiennent des valeurs nulles qui sont : counter_number , reading_remarque qui ont
pour données respectives 48 , 4531 

```python
import seaborn as sns
import matplotlib.pyplot as plt

cols = df.select_dtypes(include=['float','int']).columns
for col in cols :
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col].dropna(), bins=30, kde=True)
    plt.xlabel("Valeurs")
    plt.ylabel("Fréquence")
    plt.title(f"Histogramme et courbe de densité de {col}")
    plt.show()
```

```python
df.isnull().mean()
```

```python

for col in df.select_dtypes(include=['float', 'int']).columns:
    skewness = df[col].skew()
    print(f"Asymétrie de {col}: {skewness}")
```

Après avoir calculer l'asymetrie des differentes colonnes , nous pouvons dire ici pour les colonnes : counter_number et reading_remarque , il s'agit 
d'une distribution asymetrique  , alors nous allons utiliser la mediane pour remplacer les valeurs manquantes .

```python
df = df.fillna(df["counter_number"].median())
df= df.fillna(df["reading_remarque"].median())
```
